```python
import numpy as np
import pandas as pd
import csv

reviews = pd.read_csv("/Users/arjun/Downloads/AirBnb/Test/reviews_filtered_NYC.csv",encoding="latin1")
reviews['sentiment_label']=0
reviews['comments_matrix']=0
reviews.dropna(how='any')
#reviews=reviews.head(10)
reviews
```

```python
from nltk.sentiment.vader import SentimentIntensityAnalyzer
sid = SentimentIntensityAnalyzer()
label=[]
for sentence in reviews['comments'].values:
      ## print(sentence)\n"
    if type(sentence)==str:
        ss = sid.polarity_scores(sentence)
        if(ss['neg']>ss['pos']):
            label.append("negative")
        else:
            label.append("positive")
    else:
        label.append("NA")    
```

```python
reviews['sentiment_label']=label
df=pd.concat([reviews['comments'], reviews['sentiment_label']], axis=1, keys=['comments', 'sentiment_label'])
filename = '/Users/arjun/Downloads/AirBnb/Test/reviews_NYC_sentiment.csv'
df.to_csv(filename, index=False, encoding='utf-8')
```
